```python
import pandas as pd
```

```python
df=pd.read_csv('/content/SMSSpamCollection.txt',names=['Label','Message'],delimiter='\t',)
df.head()
```

```python
#checking for null values
df.isnull().sum()
```

```python
df.shape
```

```python
import nltk
nltk.download('stopwords')
```

```python
nltk.download('wordnet')
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
wnl=WordNetLemmatizer()
corpus=[]
for i in range(len(df)):
  words=re.sub('[^a-zA-z]',' ',df['Message'][i])
  words=nltk.word_tokenize(words)
  words=[wnl.lemmatize(word).lower() for word in words if word not in stopwords.words('english')]
  result=' '.join(words)
  corpus.append(result)
corpus
```

# **Train Test split**

```python
dummies=pd.get_dummies(df['Label'],dtype='int',drop_first=True)
```

```python
dummies
```

```python
y=dummies
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(corpus,y,test_size=0.3,random_state=18)
len(X_train),len(X_test)
```

# **Bag of Words**

```python
nltk.download('punkt')
```

```python
import nltk
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(ngram_range=(1,2))
X_train=cv.fit_transform(X_train).toarray()
X_test=cv.transform(X_test).toarray()


```

```python
from sklearn.naive_bayes import MultinomialNB
nb=MultinomialNB()
nb.fit(X_train,y_train)
y_pred=nb.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

# Precision
report =classification_report (y_test, y_pred)
print(f"classification_report: report")

# Recall
matrix = confusion_matrix(y_test, y_pred)
print(f"confusion_matrix:matrix ")
```

# **TFIDF**

```python
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
cv=TfidfVectorizer(ngram_range=(1,2))
X_train=cv.fit_transform(X_train).toarray()
X_test=cv.transform(X_test).toarray()
from sklearn.naive_bayes import MultinomialNB
nb=MultinomialNB()
nb.fit(X_train,y_train)
y_pred=nb.predict(X_test)
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

# Precision
report =classification_report (y_test, y_pred)
print(f"classification_report: report")

# Recall
matrix = confusion_matrix(y_test, y_pred)
print(f"confusion_matrix:matrix ")
```
